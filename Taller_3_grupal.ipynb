1. Cree una imagen de un canal, es decir con una forma (alto, ancho) del tamaño que desee, donde la diagonal principal hacia arriba sea blanca y desde la diagonal hacia abajo sea negra.


```python
import numpy as N                               #Impotación y defininición libreria Numpy
import matplotlib.pyplot as pt                  #Impotación y defininición libreria matplotlib. modulopyplot
import matplotlib.image as image                


imge = N.zeros((100,100,3))                     #Declaración de matriz varible

for i in range(imge.shape[0]):                  #Recorrido de ciclo en fila de matriz
    for j in range(imge.shape[1]):              #Recorrido de ciclo en elemstos de filas de matriz
        if i < j:                               #Condicional de desarrollo de grafico
            imge[i,j] = N.array([1,1,1])       #Declaración de desarrollo de grafico en la matriz creada

pt.imshow(imge)                                
pt.show()                                       #Muestra Grafico creqa
```
